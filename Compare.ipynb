# Compare last year's data with current tweets
Necessary imports



```
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

# Parameters

```
n_account = 1000
n_tweets = 50
h_filename ='/content/drive/My Drive/accounts_updated.csv'
h_folder = '/content/drive/My Drive/csv'
h_previous = '/content/drive/My Drive/preprocessed.csv'
```

```
from google.colab import drive
drive.mount('/content/drive')
```

## Select most offensive n accounts

```
df_acc = pd.read_csv(h_filename, index_col=0)
df_acc = df_acc[df_acc['exists'] == True]
df_acc = df_acc.sort_values(by='OFF', ascending=False)[:n_account]
print(len(df_acc))
df_acc.head()
```

## Get previous tweets

```
df_prev = pd.read_csv(h_previous)
df_prev.head()
```

```
accounts = df_acc.index.tolist()
print(accounts)
```

```
print(accounts[:10])
```

```
df_prev = df_prev[df_prev.screen_name.isin(accounts)]
df_prev.head()
```

# Download previous tweets

```
os.mkdir('csv_prev_2')
```

```
count = 0
for acc in tqdm(accounts):
  df_tweet_acc = df_prev[df_prev.screen_name == acc].copy()
  if count < 10:
    if df_tweet_acc['exists'].all():
      df_tweet_acc.to_csv('csv_prev_2/' + acc + "_top10.csv")
      count = count + 1
  else:
    df_tweet_acc.to_csv('csv_prev_2/' + acc + ".csv")
```

## Get current tweets

```
df_next = pd.DataFrame()
```

```
for file in os.listdir(h_folder):
  filename = file.split('.')[0]
  if filename in accounts:
    df_append = pd.read_csv(h_folder + "/" + file)
    df_append['screen_name'] = filename
    df_next = df_next.append(df_append)
```

```
df_next.head()
```

### Drop columns to clean dataframe

```
df_prev = df_prev[['tweet', 'screen_name']]
df_next = df_next[['full_text', 'screen_name']]
df_prev = df_prev.rename(columns={'tweet': 'tweet_previous'})
df_next = df_next.rename(columns={'full_text': 'tweet_now'})
```

```
df_prev.head()
```

```
df_next.head()
```

## Compare two dataframes

```
df_both = pd.DataFrame(columns=['screen_name', 'tweet_previous', 'tweet_now'])
```

```
for acc in accounts:
  prev = df_prev[df_prev.screen_name == acc][:n_tweets]
  now = df_next[df_next.screen_name == acc][:n_tweets]
  prev['tweet_now'] = now['tweet_now'].to_numpy()
  df_both = df_both.append(prev)
```
